Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deberá ingresar estos datos. Puedes obtener más información de su cálculo en:

La función debe clasificar el resultado en sus respectivas categorías

### Solución

"Para calcular el IMC es necesario dividir el peso en kg de un paciente entre el cuadrado de su altura en metros. Con esto obtendremos un valor que se situará en uno de los siguientes rangos:

Bajo peso: menos de 18,5.
Peso normal: 18,5-24,9.
Sobrepeso: 25-29,9.

```python
peso = 0
altura = 0

peso = int(input('Introduzca su peso en kg'))
altura = float((input('Introduzca su altura en cm')))/100

IMC = round(peso/altura**2, 1)

if IMC < 18.5 : mensaje = 'Bajo peso (IMC < 18.5)'
elif 18.5 < IMC < 25 : mensaje = 'Peso normal (IMC entre 18.5 y 25)'
elif 25 < IMC < 29.9 : mensaje = 'Sobrepeso (IMC entre 25 y 30)'
else: mensaje = 'Obesidad (IMC > 30)'

print('Peso:', peso, 'kg')
print('Altura:', altura, 'm')
print('IMC:', IMC)
print('Significado:', mensaje)
```

### - Ejercicio 2
Conversor de temperaturas.

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re). Selecciona al menos 2 y construye el conversor.

```python
from IPython.display import Image
Image(filename='temp-conv-formulas.jpg')
```

```python

# Función de conversión entre ºC, ºF y ºK
def temp_convert(temp):

    valor_entrada = float(temp[:-2])

    if temp[-1] == 'C':        
        F = round(1.8*valor_entrada + 32 , 2)
        K = round(valor_entrada + 273.15 , 2)
        print(temp,' = ', F,'ºF', ' = ', K, 'K', sep = '')

    elif temp[-1] == 'K':
        C = round(valor_entrada - 273.15 , 2)
        F = round((valor_entrada - 273.15)*1.8 + 32 , 2)
        print(temp,' = ', C,'ºC', ' = ', F, 'ºF', sep = '')

    elif temp[-1] == 'F':
        C = round((valor_entrada - 32)*5/9 , 2)
        K = round((valor_entrada - 32)*5/9 +273.15 , 2)
        print(temp,' = ', C,'ºC', ' = ', K, 'K', sep = '')

# Testeo de la función
for temp in ('0ºC','0ºF','0ºK', '100ºC','100ºF','100ºK'):
    temp_convert(temp)
```

### - Ejercicio 3
Contador de palabras de un texto.

Escribe una función que dado un texto, muestre las veces que aparece cada palabra.

```python

# Función que cuenta las palabras de un texto

def contar_palabras(texto):

    # caracteres para eliminar del texto para evitar que modifiquen las palabras
    caracteres_especiales = ('.', ',', '"', ':', ';', '?', '!', '-', '()', ')')

    # Comprobación caracter por caracter si está en la lista de caracteres especiales
    # Si no lo está se convierte en minúscula con lower()
    # El texto se restaura con ''.join() y se convierte en lista con split()

    palabras = ''.join([caracter.lower() for caracter in texto if caracter not in caracteres_especiales]).split()

    # Diccionario para el recuento de las palabras
    recuento_palabras = {}

    # Cada palabra se comprueba si está en la lista
    # Si está se suma uno a su cuenta, y si no está se añade con cuenta 1
    for palabra in palabras:
        if palabra in recuento_palabras:
            recuento_palabras[palabra] += 1
        else: 
            recuento_palabras[palabra] = 1

    # Se ordena el diccionario por orden descendente de sus items y pasa a ser una lista de tuplas
    recuento_palabras = sorted(recuento_palabras.items(), key=lambda x: x[1], reverse=True)

    # Si se desea ordenar por orden alfabético de las palabras, se puede omitir key=lambda x: x[0], reverse=False
    # recuento_palabras = sorted(recuento_palabras.items())

    for palabra, cuenta in recuento_palabras:
        print(palabra, ': ', cuenta, sep = '')

    

# Test con un texto cuyas palabras van a ser contadas
texto = 'El cliente quedó contento con el contador ? ! de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a la Z. Por ejemplo, para el archivo "tu_me_quieres_blanca.txt" la salida esperada sería:'

contar_palabras(texto)

```

### - Ejercicio 4
Diccionario inverso.

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiados las claves y valores. Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia.




```python
encuesta = {
    "¿Cuál es tu color favorito?": "Blanco",
    "¿Cuál es tu animal favorito?": "La ardilla",
    "¿Cuál es tu hobby principal?": "El fútbol",
    "¿Cuál es tu comida favorita?": "La fabada",
    "¿Cuál es tu género musical favorito?": "El punk",
    "¿Cuál es tu deporte favorito?": "El fútbol",
    "¿Cuál es tu película favorita?": "Harry Potter",
    "¿Cuál es tu lugar de vacaciones favorito?": "Playa",
    "¿Cuál es tu libro favorito?": "Harry Potter"
}

# Verificar si hay valores repetidos comparando la longitud de la lista de valores y un conjunto de los mismos
valores = list(encuesta.values())

if len(valores) != len(set(valores)):
    print("Se han encontrado respuestas iguales a distintas preguntas.")
    print("Algunas preguntas se combinarán para evitar error por claves duplicadas.")

# Nuevo diccionario para almacenar las respuestas como claves y las preguntas como valores en forma de elementos de listas
nueva_encuesta = {}

for pregunta, respuesta in encuesta.items():
    if respuesta not in nueva_encuesta:
        nueva_encuesta[respuesta] = [pregunta]
    else:
        nueva_encuesta[respuesta].append(pregunta)

nueva_encuesta
```

### Ejercicio 2
Conversión de tipos de datos

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán todos los elementos que pudieron convertirse en flotantes y la otra donde están los elementos que no pudieron convertirse. Ejemplo de la lista que recibe el cliente: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number' ), [1,2,'3','3.4']]

```python

# Instalar y/o importar biblioteca que convierte texto en número
# pip install word2number
from word2number import w2n

# Función que convierte strings en floats
def convert_to_float(element):
    try:
        return float(w2n.word_to_num(element))
    except (ValueError, TypeError):
        try:
            return float(element)
        except ValueError:
            return None

# Función que genera las dos listas deseadas
def sort_list(lst):
    lst_floats = []
    lst_nonfloats = []

    for elemento in lst:

        if isinstance(elemento, (list, tuple)):  # Caso de lista o tupla dentro de la lista
            for sub_elemento in elemento:
                result = convert_to_float(sub_elemento)
                if result is not None:
                    lst_floats.append(result)       # Anexar el número convertido
                else:
                    lst_nonfloats.append(sub_elemento)  # Anexar el elemento original

        else:    # Caso de texto que no está ni en una lista ni en una tupla
            result = convert_to_float(elemento)  
            if result is not None:  # Caso elemento que se ha podido convertir a float
                lst_floats.append(result)
            else:    # Caso elemento que no se ha podido convertir a float
                lst_nonfloats.append(elemento)  

    print(lst_floats)
    print(lst_nonfloats)



# Test de las funciones
lst = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

sort_list(lst)


```
